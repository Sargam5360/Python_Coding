```python
#list class name and arguments
 
import pandas as pd
#Strip Function
def extractBTEQ(line):
    if len(line) == 0:
        return
    else:
        return(line[13:-3])



def read_file(filename,l):
    with open(filename) as file:
        try:
            for line in file:
                if line.strip().startswith("self.BTEQ"):
                    l.append(extractBTEQ(line))                 
                pass
        except:
            print("exception  occured : " +str(filename))
            pass




#read the file and print
#loop and create list of lists
file = open('pathfilesampler1.txt','r')
lines = file.readlines()
listoflist = []
filepath = []
lcount = []
for line in lines:
#filename = "d3\ABD\Load\ABDLoad.py"
        #print(line)
        filename = str(line.strip('\n'))
        filepath.append(filename)
        l=[]
        read_file(filename,l)

        #print("String is:")
        #print(l)
        listoflist.append(l)
        lcount.append(len(l))
#print(listoflist)
```

```python
print(listoflist)
```

```python
d = {'FilePath':filepath,'BTEQList':listoflist, 'CountofBTEQ': lcount}
df = pd.DataFrame(d)
df.head()
```
